```python
df["mean_score"]=(df["math score"]+df["reading score"]+df["writing score"])/3
df.head()
```

```python
df.group.value_counts().plot(kind ="pie",figsize=(8,6),autopct ="%1.2f%%")
plt.axis("equal")
```

```python
g=df.sort_values(['mean_score'],ascending=False).groupby('group', as_index=False).mean()
g
```

```python
g.plot(kind="bar",x="group")
```

```python
sns.countplot("group",data=g)
```

```python
sns.pairplot(g,hue="group")
```

```python
g=df.sort_values(['mean_score'],ascending=False).groupby('gender', as_index=False).mean()
g
```
